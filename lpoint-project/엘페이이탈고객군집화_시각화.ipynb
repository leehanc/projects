## 가. 이탈고객 군집 MYSQL에서 불러오기

## 1. 이탈_군집0
###    - 요일
###    - 채널
###    - 제휴사

```python
import pandas as pd
import pymysql
conn = pymysql.connect(host='127.0.0.1', user='root', password='lpoint', db='lpoint', charset='utf8')
cur = conn.cursor()
cur.execute("select de_day, count(*) from 엘페이이탈고객군집화 a inner join 엘페이이용_view b on a.cust=b.cust where 군집=0 group by b.de_day;")
result = cur.fetchall()
conn.commit()
conn.close()
df_w = pd.DataFrame(result)
df_w.columns=['요일','이용량']
df_w=df_w.set_index('요일')

weekday = ['월요일', '화요일', '수요일', '목요일', '금요일', '토요일', '일요일']
df_w= df_w.loc[weekday]

df_w
```

```python
import pandas as pd
import pymysql
conn = pymysql.connect(host='127.0.0.1', user='root', password='lpoint', db='lpoint', charset='utf8')
cur = conn.cursor()
cur.execute("select chnl_dv, count(*) from 엘페이이탈고객군집화 a inner join 엘페이이용_view b on a.cust=b.cust where 군집=0 group by chnl_dv order by chnl_dv;")
result = cur.fetchall()
conn.commit()
conn.close()
df_c = pd.DataFrame(result)
df_c.columns=['채널','이용량']
df_c
```

```python
import pandas as pd
import pymysql
conn = pymysql.connect(host='127.0.0.1', user='root', password='lpoint', db='lpoint', charset='utf8')
cur = conn.cursor()
cur.execute("select 군집, cop_c, count(*) from 엘페이이탈고객군집화 a inner join 엘페이이용_view b on a.cust = b.cust where 군집=0 group by cop_c order by cop_c;")
result = cur.fetchall()
conn.commit()
conn.close()
df = pd.DataFrame(result)
df.columns = ['군집', '제휴사', '수']
df
```
